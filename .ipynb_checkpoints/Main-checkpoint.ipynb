# Scrapping data from wikipedia largest comapnies

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
base_url="https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue"
```

```python
response=requests.get(base_url)
```

```python
soup=BeautifulSoup(response.text,'html.parser')
```

```python
table=soup.find('table')
table
```

```python
titles=table.find_all('th')[:7]
rev=table.find_all('th')[9]
titles.append(rev)
titles
```

```python
world_title=[title.text.strip() for title in titles]
world_title
```

```python
df=pd.DataFrame(columns=world_title[1:])
df
```

```python
column_data=table.find_all('tr')
len(column_data)
```

```python
Ranks=table.find_all('th')[11:]

Rank=[rank.text.strip() for rank in Ranks]
Rank
```

```python
for row in column_data[2:]:
    row_data=row.find_all('td')
    
    single_row_data=[data.text.strip() for data in row_data]
    
    first=single_row_data[0:6]
    rev=single_row_data[8:9]
    
    combined=first+rev
    
    length=len(df)
    df.loc[length]=combined 
```

```python
df['Rank']=Rank
```

```python
df.head()
```

```python
df = df[['Rank'] + [col for col in df.columns if col != 'Rank']]
df
```

```python
df.to_csv('companies.csv',index=False)
```

# Scrapping data of BA reviews from airline quality

```python
base_url='https://www.airlinequality.com/airline-reviews/british-airways/'
pages=50
page_size=50
```

```python
reviews=[]
ratings=[]
for i in range(1,pages+1):
    url=f"{base_url}page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"
    print('Scrapping page',i)
    response=requests.get(url)
    
    soup=BeautifulSoup(response.text,'html.parser')
    
    for review in soup.find_all("div",{"class":"text_content"}):
        reviews.append(review.text)
```

```python
df=pd.DataFrame(reviews,columns=['Reviews'])
```

```python
df.to_csv("BA Reviews")
```
